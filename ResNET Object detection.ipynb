```python
from tensorflow.keras.applications.resnet50 import ResNet50,preprocess_input,decode_predictions
```

```python
from tensorflow.keras.preprocessing import image
```

```python
import numpy as np
```

```python
img_path_list=['war_plane.jpg','kaan.jpg']
```

```python
from IPython.display import Image
Image(img_path_list[0])
```

```python
Image(img_path_list[1])
```

```python
model=ResNet50(weights='imagenet')
```

```python
def img_preprocessing(image_path_):
    img=image.load_img(image_path_, target_size=(224,224))
    img=image.img_to_array(img)
    img=np.expand_dims(img,axis=0)
    img=preprocess_input(img)
    return img
```

```python
for image_path in img_path_list:
    img2=img_preprocessing(image_path)
    pred=model.predict(img2)
    print (image_path,decode_predictions(pred,top=1))
```
